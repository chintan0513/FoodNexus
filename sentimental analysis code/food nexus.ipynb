```python
import csv
from textblob import TextBlob
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python


# Function to remove stop words and tokenize a tweet
def clean_and_tokenize(tweet):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(tweet)
    filtered_tweet = [word for word in word_tokens if word.lower() not in stop_words]
    return filtered_tweet

# Function to perform sentiment analysis on a tokenized tweet
def analyze_sentiment(tokens):
    cleaned_tweet = ' '.join(tokens)  # Convert tokenized tweet back to string
    analysis = TextBlob(cleaned_tweet)
    # Determine sentiment polarity (positive, negative, neutral)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
# Read tweet data from CSV file
tweets = []
with open('dummy_tweets.csv', 'r', newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        tweets.append(row)

# Perform sentiment analysis after removing stop words and tokenizing each tweet
for tweet in tweets:
    cleaned_tokens = clean_and_tokenize(tweet['Tweet'])
    tweet['Sentiment'] = analyze_sentiment(cleaned_tokens)

# Write the cleaned data with sentiment analysis results to a new CSV file
headers = ['Timestamp','Location', 'Continent', 'Tweet', 'Day', 'Date', 'Time', 'Year', 'Sentiment']
with open('cleaned_data_without_algorithm.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=headers)
    writer.writeheader()
    for tweet in tweets:
        writer.writerow(tweet)

print("Cleaned data saved to 'cleaned_data_without_algorithm.csv'")
```

# Cross validation

```python
from sklearn.model_selection import train_test_split

# Split data into training and testing sets
train_tweets, test_tweets = train_test_split(tweets, test_size=0.2, random_state=42)

# Train TextBlob sentiment analysis on the training set
for tweet in train_tweets:
    cleaned_tokens = clean_and_tokenize(tweet['Tweet'])
    tweet['Sentiment'] = analyze_sentiment(cleaned_tokens)

# Evaluate TextBlob sentiment analysis on the testing set
correct_predictions = 0
total_tweets = len(test_tweets)

for tweet in test_tweets:
    cleaned_tokens = clean_and_tokenize(tweet['Tweet'])
    predicted_sentiment = analyze_sentiment(cleaned_tokens)
    tweet['Predicted_Sentiment'] = predicted_sentiment
    
    # Compare TextBlob's prediction with the true sentiment
    if predicted_sentiment == tweet['Sentiment']:
        correct_predictions += 1

# Calculate accuracy
accuracy = correct_predictions / total_tweets
print("Cross-Validation Accuracy:", accuracy)
```

# K-Fold Cross-Validation

```python
from sklearn.model_selection import KFold

# Perform k-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
accuracies = []

for train_index, test_index in kf.split(tweets):
    train_tweets = [tweets[i] for i in train_index]
    test_tweets = [tweets[i] for i in test_index]

    # Train TextBlob sentiment analysis on the training set
    for tweet in train_tweets:
        cleaned_tokens = clean_and_tokenize(tweet['Tweet'])
        tweet['Sentiment'] = analyze_sentiment(cleaned_tokens)

    # Evaluate TextBlob sentiment analysis on the testing set
    correct_predictions = 0
    total_tweets = len(test_tweets)

    for tweet in test_tweets:
        cleaned_tokens = clean_and_tokenize(tweet['Tweet'])
        predicted_sentiment = analyze_sentiment(cleaned_tokens)
        tweet['Predicted_Sentiment'] = predicted_sentiment
        
        # Compare TextBlob's prediction with the true sentiment
        if predicted_sentiment == tweet['Sentiment']:
            correct_predictions += 1

    # Calculate accuracy for this fold
    accuracy = correct_predictions / total_tweets
    accuracies.append(accuracy)

# Calculate average accuracy across all folds
avg_accuracy = sum(accuracies) / len(accuracies)
print("K-Fold Cross-Validation Average Accuracy:", avg_accuracy)
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import csv

# Download NLTK resources if not already downloaded
nltk.download('vader_lexicon')
nltk.download('punkt')
nltk.download('stopwords')

# Initialize VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Function to perform document-level sentiment analysis using VADER
def analyze_document_sentiment_vader(tweet):
    sentiment_scores = sid.polarity_scores(tweet)
    compound_score = sentiment_scores['compound']
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Function to perform sentence-level sentiment analysis using VADER
def analyze_sentence_sentiment_vader(tweet):
    sentences = sent_tokenize(tweet)
    sentence_sentiments = []
    for sentence in sentences:
        sentiment_scores = sid.polarity_scores(sentence)
        compound_score = sentiment_scores['compound']
        if compound_score >= 0.05:
            sentence_sentiments.append('Positive')
        elif compound_score <= -0.05:
            sentence_sentiments.append('Negative')
        else:
            sentence_sentiments.append('Neutral')
    return sentence_sentiments

# Function to perform aspect-level sentiment analysis using VADER
def analyze_aspect_sentiment_vader(tweet, aspects):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(tweet)
    filtered_tokens = [word for word in word_tokens if word.lower() not in stop_words]
    aspect_sentiments = {}
    for aspect in aspects:
        if aspect in filtered_tokens:
            aspect_sentiments[aspect] = analyze_document_sentiment_vader(aspect)
        else:
            aspect_sentiments[aspect] = 'Neutral'
    return aspect_sentiments

# Read tweet data from CSV file
tweets = []
with open('dummy_tweets_modified.csv', 'r', newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        tweets.append(row)

# Define aspects related to food waste and poverty
aspects = [
    "Hunger",
    "Poverty",
    "Malnutrition",
    "Food insecurity",
    "Homelessness",
    "Aid",
    "Charity",
    "Unemployment",
    "Economic hardship",
    "Social justice"
]


# Perform sentiment analysis at different levels
for tweet in tweets:
    tweet['Document_Sentiment'] = analyze_document_sentiment_vader(tweet['Tweet'])
    tweet['Sentence_Sentiments'] = analyze_sentence_sentiment_vader(tweet['Tweet'])
    tweet['Aspect_Sentiments'] = analyze_aspect_sentiment_vader(tweet['Tweet'], aspects)

# Write the cleaned data with sentiment analysis results to a new CSV file
headers = ['Location', 'Continent', 'Tweet', 'Day', 'Date', 'Time', 'Year', 'Document_Sentiment', 'Sentence_Sentiments', 'Aspect_Sentiments']
with open('combined_sentiment_analysis.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=headers)
    writer.writeheader()
    for tweet in tweets:
        writer.writerow(tweet)

print("Combined sentiment analysis results saved to 'combined_sentiment_analysis.csv'")
```
